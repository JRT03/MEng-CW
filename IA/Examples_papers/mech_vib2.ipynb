```python
# q10

import numpy as np
import matplotlib.pyplot as plt
#Defining perameters
omn = 100*2*np.pi
damp = 0.03
overtime = 4
timestep = 0.0003

#Creating discrete time steps
time = np.arange(0, overtime, timestep)

#Initalising variables

ydd = np.zeros(len(time))
yd = np.zeros(len(time))
y = np.zeros(len(time))
x = np.zeros(len(time))

#inputing starting conditions
x[0] = 1
y[0] = 0
yd[0] = 0

#iterating through entire time length
for i in range(1, len(time)-1):    
    ydd[i] = (omn**2) * ( x[i-1] - 2*damp*yd[i-1]/omn - y[i-1] )
    yd[i] = yd[i-1] + ydd[i]*timestep
    y[i] = y[i-1] + yd[i]*timestep

plt.plot(time,y)
plt.xlim(0, 0.5)
plt.show()

```

```python


#Defining perameters
omn = 100*2*np.pi
damp = 0.03
overtime = 0.4
timestep = 0.0003
Amplitude = 0.5
f = 150

#Creating discrete time steps
time = np.arange(0, overtime, timestep)
print(time)

y = omn*(np.e**(-1*damp*omn*time))*np.sin(omn*time)

plt.plot(time,y)

#sinusoidal input
ip = Amplitude*np.sin(f*time)

g = np.convolve(y,ip,mode='same')

plt.plot(time,g)

plt.show()
```
